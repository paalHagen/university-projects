```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.karate_club_graph()

# Task 1.1
print("Nodes with attributes")
print(G.nodes(data=True))
print("Edges with attributes")
print(G.edges(data=True))

# Task 1.2
colors = []
for node in G.nodes:
    if G.nodes[node]["club"] == 'Mr. Hi':
        colors.append("red")
    else:
        colors.append("blue")

plt.title(label="Task 1.2")
nx.draw_circular(G, with_labels=True, node_color=colors)
plt.show()

# Task 1.3
short_path = nx.dijkstra_path(G, source=24, target=16)
print("List of nodes:", short_path)

# Task 1.4
colors_2 = []
for node in G.nodes:
    if node in short_path:
        colors_2.append("red")
    else:
        colors_2.append("blue")

plt.title(label="Task 1.4")
nx.draw_circular(G, with_labels=True, node_color=colors_2)
plt.show()
```
